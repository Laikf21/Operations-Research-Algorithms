# 3 PoolSearchMode设置

## 3.1 PoolSearchMode = 0, MIP追求寻找最优解

```python
model = gp.read("VRPTW_r102_20_5.mps")

model.Params.LogToConsole = 1

model.Params.PoolSolutions = 20

# PoolSearch属性设置求解目标，0代表只注重找到最优解
model.Params.PoolSearchMode = 0

model.optimize()

# for i in range(10):
#     model.setParam(gp.GRB.Param.SolutionNumber, i)
#     print("Obj_{} = {}" .format(i+1, model.PoolObjVal))
```

## 3.2 PoolSearchMode = 1, MIP求解器会寻找尽可能多的解，但不保证解的质量。

```python
model = gp.read("VRPTW_r102_20_5.mps")

model.Params.LogToConsole = 1

model.Params.PoolSolutions = 20

# PoolSearch属性设置求解目标，1代表找到尽可能多的解，但不保证解的质量
model.Params.PoolSearchMode = 1

model.optimize()

# for i in range(10):
#     model.setParam(gp.GRB.Param.SolutionNumber, i)
#     print("Obj_{} = {}" .format(i+1, model.PoolObjVal))
```

## 3.3 PoolSearchMode = 2, MIP会尽可能找到PoolSolutions参数所要求的n个解，同时尽可能保证它们为n个最优解。

```python
model = gp.read("VRPTW_r102_20_5.mps")
model.Params.LogToConsole = 1

model.Params.PoolSolutions = 20

# PoolSearch属性设置求解目标，1代表找到尽可能多的解，但不保证解的质量
model.Params.PoolSearchMode = 2

model.optimize()

# for i in range(10):
#     model.setParam(gp.GRB.Param.SolutionNumber, i)
#     print("Obj_{} = {}" .format(i+1, model.PoolObjVal))
```

# 4 PoolGap 参数

```python
model = gp.read("VRPTW_r102_20_5.mps")
model.Params.LogToConsole = 1

model.Params.PoolSolutions = 10

#求解器将舍弃不满足PoolGap参数所设定Gap条件的解
model.Params.PoolGap = 0.1

model.optimize()

for i in range(model.SolCount):
    model.Params.SolutionNumber = i
    print("Obj_{} = {}" .format(i+1, model.PoolObjVal))
```
